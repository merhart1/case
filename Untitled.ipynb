```python
f = Figure( figsize =(7,7) )
#Corr plot int_rate v. annual_inc
g = sns.jointplot(x="int_rate", y="annual_inc", data=newLoanDF, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
```
